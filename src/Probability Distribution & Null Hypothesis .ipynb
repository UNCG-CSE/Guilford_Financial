### One-Sample T-Test

- whether the average money spent in 2013 differs from the population:
    - Null Hypethoesis H0 - 'Total Expenses in 2013 is similar to all expenses'
    - Alternative Hypothesis H1 - 'Total Expenses in 2013 is different to all expenses'

```python
data
```

```python
df_transposed.groupby(['Year'])['Public Health'].mean()
```

```python
population_mean = df_transposed['Public Health'].mean()
population_mean
```

```python
# data_2013 = data.groupby(by=['Year'])
```

```python
data = data[data['Dept Name'] == 'Public Health']
# data.iloc[:,16]

data
```

```python
data = data[data['Year'] == 2013].Amount
```

```python
stats.ttest_1samp(data, population_mean)
```

- The test result shows
    - the test statstic "t" is euqal to -1.515
    - the p-value is 0.130 (> 0.05)

```python
data.shape[0]
```

```python
stats.t.ppf(q=0.025, df=1348)
```

```python
stats.t.cdf(x=-1.5151,
            df=1348) * 2
```

Notice this value is the same as the p-value listed in the original t-test output. 
- A p-value of 0.130 means we'd expect to see data as extreme as our sample due to chance about 13% of the time if the null hypothesis was true. 
<br></br><br></br>
    - In this case, the p-value is lower than our significance level α (equal to 1-conf.level or 0.05) so, ***we should not reject the null hypothesis.***
