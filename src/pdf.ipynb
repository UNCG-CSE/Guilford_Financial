```python
############PDFDFDFDFDF#33333333

## remove negative values ###
Expense_Budget_positive=Expense_Budget[Expense_Budget.Amount >0]
```

```python
Expense_Budget_positive.groupby("Year").Amount.plot(kind='density',figsize=(8,8))
```

```python
Expense_Budget_positive.groupby("Year").Amount.count()
```

```python
alpha_mom = Expense_Budget_positive.groupby("Year").Amount.mean() ** 2 / Expense_Budget_positive.groupby("Year").Amount.var()
beta_mom = Expense_Budget_positive.groupby("Year").Amount.var() / Expense_Budget_positive.groupby("Year").Amount.mean()

from scipy.stats.distributions import gamma

Expense_Budget_positive.groupby("Year").Amount.hist(density=True, bins=5, ec='black')
label = 'alpha = {0:.2f}\nbeta = {1:.2f}'.format(alpha_mom[0], beta_mom[0])
plt.annotate(label, xy=(10, 0.2))
plt.plot(np.linspace(0, 10), gamma.pdf(np.linspace(0, 10), alpha_mom[0], beta_mom[0]))
```
