```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib
```

```python
data = pd.read_csv('../data/Expense_Budget.csv')
```

```python
data.head()
```

```python
data['Dept Name'].value_counts()
```

```python
x = list(data['Dept Name'].unique())
```

```python
y1 = data.groupby(['Year','Dept Name'])['Amount'].mean().loc[2013]
y2 = data.groupby(['Year','Dept Name'])['Amount'].mean().loc[2014]
y3 = data.groupby(['Year','Dept Name'])['Amount'].mean().loc[2015]
y4 = data.groupby(['Year','Dept Name'])['Amount'].mean().loc[2016]
y5 = data.groupby(['Year','Dept Name'])['Amount'].mean().loc[2017]
y6 = data.groupby(['Year','Dept Name'])['Amount'].mean().loc[2018]
df = pd.DataFrame({'2013': y1, '2014': y2, '2015': y3, '2016': y4, '2017': y5, '2018': y6})
```

```python
plt.figure(figsize=(15,10))
plt.xlabel('Department Name')
plt.ylabel('Expense')
plt.title('Expenses for Different Department')
plt.plot(x, y1,'r', data=data,label='2013')
plt.plot(x, y2,'g', data=data,label='2014')
plt.plot(x, y3,'y', data=data,label='2015')
plt.plot(x, y4,'orange', data=data,label='2016')
plt.plot(x, y5,'b', data=data,label='2017')
plt.plot(x, y6,'purple', data=data,label='2018')
plt.xticks(rotation=90)
plt.legend()
# plt.savefig("trends.png")
```

```python
y1.plot(kind="density", figsize=(10,10), xlim=(-1.0* 10**8, 3* 10**8), label='2013')
y2.plot(kind="density", figsize=(10,10), xlim=(-1.0* 10**8, 3* 10**8), label='2014')
y3.plot(kind="density", figsize=(10,10), xlim=(-1.0* 10**8, 3* 10**8), label='2015')
y4.plot(kind="density", figsize=(10,10), xlim=(-1.0* 10**8, 3* 10**8), label='2016')
y5.plot(kind="density", figsize=(10,10), xlim=(-1.0* 10**8, 3* 10**8), label='2017')
y6.plot(kind="density", figsize=(10,10), xlim=(-1.0* 10**8, 3* 10**8), label='2018')
plt.legend()
plt.xlabel('Expense')
plt.ylabel('Density')
plt.title('Density Graph of Total Expenses for Different Years')
plt.savefig("peak")
```
