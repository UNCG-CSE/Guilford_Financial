# Anomaly Detection Using IQR

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set some Pandas options
pd.set_option('notebook_repr_html', False)
pd.set_option('max_columns', 30)
pd.set_option('max_rows', 20)
```

## Expense Anomalies

```python
expbudg = pd.read_csv("C:\Users\cothe\Desktop\CSC\CSC505\FinalProject\Guilford_Financial\data\Expense_Budget.csv")
expbudg.head()
```

### Number of Anomalies per Department

```python
exp_group = expbudg.groupby(by=["Dept Name"])
expdeptlist = []
for dept, group in exp_group:
    expAnomalyCount = 0
    q1 = np.percentile(group.Amount, 25)
    q3 = np.percentile(group.Amount, 75)
    iqr = q3 - q1
    for amt in group.Amount:
        if (((amt < (q1 - (3*iqr))))) or (((amt > (q3 + (3*iqr))))):
            expAnomalyCount += 1
    print dept
    print expAnomalyCount
    x = dept, expAnomalyCount
    expdeptlist.append(x)
```

#### Barplot

```python
deptAnom = pd.Series()
for y in expdeptlist:
    deptAnom.set_value(y[0],y[1])
da = deptAnom.plot(kind='bar', figsize=(20,10), title = 'Number of Anomalies in Expense per Department Over 5 Years')
da.set_xlabel("Department Names")
da.set_ylabel("Number of Anomalies")
```

#### Histogram

```python
da = deptAnom.hist(figsize=(20,10), bins = 10)
da.set_xlabel("Number of Anomalies")
da.set_ylabel("Probability")
da.set_title("All Time Anomaly Probability")
```

### Number of Anomalies per Department per Year

```python
exp_group = expbudg.groupby(by=["Dept Name"])
expdeptlist = []
for dept, group in exp_group:
    total = 0
    yGroup = group.groupby(by=["Year"])
    print dept
    for year, gp in yGroup:
        expAnomalyCount = 0
        q1 = np.percentile(group.Amount, 25)
        q3 = np.percentile(group.Amount, 75)
        iqr = q3 - q1
        for amt in gp.Amount:
            if (((amt < (q1 - (3*iqr))))) or (((amt > (q3 + (3*iqr))))):
                expAnomalyCount += 1
                total +=1
        print year
        print expAnomalyCount
        x = dept, expAnomalyCount, year
        expdeptlist.append(x)
    print "Total: " + str(total)
```
