```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Set some Pandas options
pd.set_option('notebook_repr_html', False)
pd.set_option('max_columns', 30)
pd.set_option('max_rows', 20)
```

```python
expbudg = pd.read_csv("C:\Users\cothe\Desktop\CSC\CSC505\FinalProject\Guilford_Financial\data\Expense_Budget.csv")
expbudg.head()
```

Number of Anomalies Per Department

```python
#fig, ax = plt.subplots()
#ind = np.arange(1, 4)

exp_group = expbudg.groupby(by=["Dept Name"])
expdeptlist = []
for dept, group in exp_group:
    expAnomalyCount = 0
    i = 0
    for amt in group.Amount:
        if ((amt - group.Amount.mean()) > 3 * group.Amount.std()):
            expAnomalyCount += 1
    #print dept
    #print expAnomalyCount
    x = dept, expAnomalyCount
    expdeptlist.append(x)

deptAnom = pd.Series()
for y in expdeptlist:
    deptAnom.set_value(y[0],y[1])
da = deptAnom.plot(kind='bar', figsize=(20,10), title = 'Number of Anomalies in Expense per Department Over 5 Years')
da.set_xlabel("Department Names")
da.set_ylabel("Number of Anomalies")
```
